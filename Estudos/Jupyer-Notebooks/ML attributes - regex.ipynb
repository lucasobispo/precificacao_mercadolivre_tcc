```python
url = f'https://api.mercadolibre.com/sites/MLB/search?category=MLB1051&since=today'

r = urllib2.Request(url)
response = urllib2.urlopen(r)
elevations = response.read()
data = json.loads(elevations)
df = pd.json_normalize(data['results'])
```

```python
teste = df.attributes[0]
```

```python
dic_flattened = [flatten(d) for d in data['results']]
df = pd.DataFrame(dic_flattened)
```

```python
df.head(5)
```

```python
df['title_upper'] = df['title'].str.upper()
```

```python
GB_search = []
GBRAM_search = []
for values in df['title_upper']:
    
    if re.findall(r"(\d+) GB(?!RAM|\.).*", values):
        GB_search.append(*re.findall(r"(\d+) GB(?!RAM|\.).*", values))
    else:
        GB_search.append(None)
    
    if re.findall(r"(\d+) GB RAM", values):
        GBRAM_search.append(*re.findall(r"(\d+) GB RAM", values))
    else:
        GBRAM_search.append(None)
    
    print(values)
    print(*GB_search)
df['memory'] = GB_search
df['ram_memory'] = GBRAM_search
```
